# importamos libreria

```python
import pandas as pd
```

# cargamos el dataset

```python
df = pd.read_csv("xdxdxd - 2022-08-23_CasosConfirmados.csv")
df
```

# separamos las columnas en x e y

```python
colnames = df.columns.values.tolist()
predictors = colnames[:13]
target = colnames[13]
X = df[predictors]
Y = df[target]
```

# importamos la regresión de arboles de decisión

```python
from sklearn.tree import DecisionTreeRegressor
```

# aplicamos la regresión 

```python
regtree = DecisionTreeRegressor(min_samples_split=30, min_samples_leaf=10, max_depth=5, random_state=0)
```

```python
regtree.fit(X,Y)
```

# creamos la predicción de los fallecidos 

```python
preds = regtree.predict(df[predictors])
```

```python
df["preds"] = preds
```
